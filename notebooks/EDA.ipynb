```python
#Dividing past 12 month consumption column into different consumption subgroups to analyze the mean churn and consumer groupings.

bins = [0, 200000, 400000, 600000, client_data['cons_12m'].max()]
labels = ['0-200k', '200k-400k', '400k-600k', '600k+']

client_data['cons_12m_group'] = pd.cut(
    client_data['cons_12m'],
    bins=bins,
    labels=labels,
    include_lowest=True
)
```

```python
grouped = client_data.groupby(['cons_12m_group', 'churn'])['id'].count().unstack().fillna(0)
grouped_percentage = grouped.div(grouped.sum(axis=1), axis=0) * 100

print(grouped_percentage)


ax = grouped_percentage.plot(kind='bar', stacked=True, figsize=(8,6), colormap='Set2')
plt.title('Churn vs Not churned consumers by consumption group')
plt.xlabel('Consumption Group (past 12 months)')
plt.ylabel('Percentage of Customers')
plt.legend(['Not Churned', 'Churned'], title = 'Churn Status')


for i, group in enumerate(grouped_percentage.index):
    bottom = 0
    for j, status in enumerate(grouped_percentage.columns):
        percent = grouped_percentage.loc[group, status]
        if percent > 0:
            ax.text(i, bottom + percent/2, f'{percent:.1f}%', ha = 'center', va = 'center', fontsize = 9, color = 'black')
        bottom += percent



plt.tight_layout()
plt.show()
```

```python
grouped_percentage_renamed = grouped_percentage.rename(columns={0: 'Not Churned', 1: 'Churned'})
grouped_renamed = grouped.rename(columns={0: 'Not Churned', 1: 'Churned'})

fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(14, 6))


grouped_percentage_renamed.plot(kind='bar', stacked=True, ax=axs[0], colormap='Set2')
axs[0].set_title('Churn Percentage by Consumption Group')
axs[0].set_ylabel('Percentage')
axs[0].set_xlabel('Consumption Group')


for i, group in enumerate(grouped_percentage_renamed.index):
    bottom = 0
    for status in ['Not Churned', 'Churned']:
        percent = grouped_percentage_renamed.loc[group, status]
        axs[0].text(i, bottom + percent / 2, f'{percent:.1f}%', ha='center', va='center', fontsize=9, color='black')
        bottom += percent


grouped_renamed.plot(kind='bar', stacked=True, ax=axs[1], colormap='Set2')
axs[1].set_title('Churn Count by Consumption Group')
axs[1].set_ylabel('Count')
axs[1].set_xlabel('Consumption Group')


for i, group in enumerate(grouped_renamed.index):
    bottom = 0
    for status in ['Not Churned', 'Churned']:
        count = grouped_renamed.loc[group, status]
        axs[1].text(i, bottom + count / 2, int(count), ha='center', va='center', fontsize=9, color='black')
        bottom += count

plt.tight_layout()
plt.show()
```

# Visualizing churn among different groups.

```python
def plot_stacked_bars(df, title_, size = (18,10), rot_= 0, legend_='upper right'):
    """
    Plots a stacked bar chart with annotations.
    
    """
    ax = df.plot(kind='bar', stacked = True, figsize=size, rot=rot_, title=title_)

    annotate_stacked_bars(ax, textsize=14)

    plt.legend(["Not Churned", "Churned"], loc=legend_)
    plt.ylabel("Company base (%)")
    plt.show()

def annotate_stacked_bars(ax, pad=0.99, color='white', textsize=13):
    """ Adding annotations to stacked bars """
    for p in ax.patches:
        
        # Calculate annotation
        value = str(round(p.get_height(),1))
        # If value is 0 do not annotate
        if value == '0.0':
            continue
        ax.annotate(
            value,
            ((p.get_x()+ p.get_width()/2)*pad-0.05, (p.get_y()+p.get_height()/2)*pad),
            color=color,
            size=textsize
        )
```

```python
#Company churn visualization
churn = client_df[['id', 'churn']]
churn.columns = ['Companies', 'Churn']
churn_total = churn.groupby(churn['Churn']).count()
churn_percentage = churn_total/churn_total.sum() * 100

plot_stacked_bars(churn_percentage.transpose(), "Churning status", (5,5), legend_="lower right")

```

# Churn by consumption

```python
consumption = client_df[['id', 'cons_12m', 'cons_gas_12m', 'cons_last_month', 'imp_cons', 'has_gas', 'churn']]
```

```python
def plot_distribution(df, column, ax, bins_=50):
    """ Plots the distribution in a stacked histogram style of churned vs retained company. """

    temp = pd.DataFrame({"Retention" : df[df['churn']==0][column],
                        "Churn":df[df['churn']==1][column]})
    
    temp[["Retention", "Churn"]].plot(kind='hist', bins=bins_, ax=ax, stacked=True)
    ax.set_xlabel(column)
    ax.ticklabel_format(style='plain', axis='x')
```

```python
fig, axs = plt.subplots(nrows=4, figsize=(18,25))



plot_distribution(consumption, 'cons_12m', axs[0])
plot_distribution(consumption[consumption['has_gas'] == 't'], 'cons_gas_12m', axs[1])
plot_distribution(consumption, 'cons_last_month', axs[2])
plot_distribution(consumption, 'imp_cons', axs[3])
```

### As evident from the visualization, the distribution of the consumption is highly skewed, implying the presence of outliers. Plotting boxplots to better check the outliers.

```python
fig, axes = plt.subplots(nrows=4, figsize=(18, 25))

sns.boxplot(x=consumption['cons_12m'], ax=axes[0])
sns.boxplot(x=consumption[consumption['has_gas'] == 't']['cons_gas_12m'], ax=axes[1])
sns.boxplot(x=consumption['cons_last_month'], ax=axes[2])
sns.boxplot(x=consumption['imp_cons'], ax=axes[3])

axs[0].set_xlim(-200000, 2000000)
axs[1].set_xlim(-200000, 2000000)
axs[2].set_xlim(-20000, 100000)
```
